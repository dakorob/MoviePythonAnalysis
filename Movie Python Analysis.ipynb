```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts confirgurations of plots

# Create dataframe

df = pd.read_csv(r'C:\Users\dakot\Downloads\movies.csv')
```

```python
# Let's take a look at the data

df.head()
```

```python
# Checking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Clearing out missing data

df = df.dropna()
```

```python
# Double-checking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Checking data types for columns

df.dtypes
```

```python
# Changing data type to int64 for relevant columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')

df['runtime'] = df['runtime'].astype('int64')
```

```python
df.head
```

```python
# Mismatch between year and released columns
# Creating year correct column using released data

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

df.head
```

```python
# Sorting by highest grossing films, descending

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Dropping duplicates

df.drop_duplicates()
```

```python
# What is correlated to highest gross revenue?

# Budget correlation - inputs?

# Company correlation - network effect?
```

```python
# Scatter plot with film budget and gross revenue

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Film Budget & Gross Revenue')

plt.xlabel('Gross Revenue')

plt.ylabel('Film Budget')

plt.show()
```

```python
df.head()
```

```python
# Plot Film Budget & Gross Revenue with seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
# Checking numeric correlation

df.corr(method='pearson') #pearson, kendall, spearman
```

```python
# Higher correlation between Film Budget & Gross Revenue
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Film Attributes')

plt.xlabel('Film Attributes')

plt.ylabel('Film Attributes')

plt.show()
```

```python
# Company correlation exploration

df.head()
```

```python
df_categorized = df

for col_name in df_categorized.columns:
    if(df_categorized[col_name].dtype == 'object'):
        df_categorized[col_name] = df_categorized[col_name].astype('category')
        df_categorized[col_name] = df_categorized[col_name].cat.codes
        
df_categorized
```

```python
correlation_matrix = df_categorized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Film Attributes')

plt.xlabel('Film Attributes')

plt.ylabel('Film Attributes')

plt.show()
```

```python
df_categorized.corr()
```

```python
correlation_mat = df_categorized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
